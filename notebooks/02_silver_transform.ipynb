```python
from pyspark.sql.functions import *
```

## read bronze tables

```python
orders = spark.read.format("delta").load(
"/Volumes/ecommerce_catalog/ecommerce_schema/ecommerce_volume/bronze/orders")

customers = spark.read.format("delta").load(
"/Volumes/ecommerce_catalog/ecommerce_schema/ecommerce_volume/bronze/customers")

products = spark.read.format("delta").load(
"/Volumes/ecommerce_catalog/ecommerce_schema/ecommerce_volume/bronze/products")

payments = spark.read.format("delta").load(
"/Volumes/ecommerce_catalog/ecommerce_schema/ecommerce_volume/bronze/payments")
```

## orders cleaning

```python
orders_clean = orders.dropna()
orders_clean = orders_clean.dropDuplicates()
orders_clean = orders_clean.withColumn("order_date",to_date("order_date"))

orders_clean.write.format("delta").mode("overwrite") \
.save("/Volumes/ecommerce_catalog/ecommerce_schema/ecommerce_volume/silver/orders")
```

## customers cleaning

```python
customers_clean = customers.dropna()
customers_clean = customers_clean.dropDuplicates()

customers_clean.write.format("delta").mode("overwrite") \
.save("/Volumes/ecommerce_catalog/ecommerce_schema/ecommerce_volume/silver/customers")
```

## products cleaning

```python
products_clean = products.dropna()
products_clean = products_clean.dropDuplicates()

products_clean.write.format("delta").mode("overwrite") \
.save("/Volumes/ecommerce_catalog/ecommerce_schema/ecommerce_volume/silver/products")
```

## payments cleaning

```python
payments_clean = payments.dropna()
payments_clean = payments_clean.dropDuplicates()

payments_clean.write.format("delta").mode("overwrite") \
.save("/Volumes/ecommerce_catalog/ecommerce_schema/ecommerce_volume/silver/payments")
```
