```python
from delta.tables import *
```

## read new orders

```python
new_orders = spark.read.csv(
"/Volumes/ecommerce_catalog/ecommerce_schema/ecommerce_volume/raw/orders_new.csv",
header=True,
inferSchema=True
)
```

## merge incremental records

```python
deltaTable = DeltaTable.forPath(spark,
"/Volumes/ecommerce_catalog/ecommerce_schema/ecommerce_volume/bronze/orders")
(
deltaTable.alias("target")
.merge(
new_orders.alias("source"),
"target.order_id = source.order_id"
)
.whenNotMatchedInsertAll()
.execute()
)
```
