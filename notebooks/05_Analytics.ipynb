## create gold tables/views

```python
fact_sales = spark.read.format("delta").load(
"/Volumes/ecommerce_catalog/ecommerce_schema/ecommerce_volume/gold/fact_sales")

dim_customer = spark.read.format("delta").load(
"/Volumes/ecommerce_catalog/ecommerce_schema/ecommerce_volume/gold/dim_customer")

dim_product = spark.read.format("delta").load(
"/Volumes/ecommerce_catalog/ecommerce_schema/ecommerce_volume/gold/dim_product")

fact_sales.createOrReplaceTempView("fact_sales")
dim_customer.createOrReplaceTempView("dim_customer")
dim_product.createOrReplaceTempView("dim_product")
```

## top products

```python
%sql
SELECT product_name,SUM(revenue) AS revenue FROM fact_sales GROUP BY product_name ORDER BY revenue DESC;
```

## revenue by state

```python
%sql
SELECT state,SUM(revenue) AS revenue FROM fact_sales GROUP BY state ORDER BY revenue DESC;
```

## monthly revenue

```python
%sql
SELECT MONTH(order_date) AS month_no,SUM(revenue) AS revenue FROM fact_sales GROUP BY MONTH(order_date) ORDER BY month_no;
```

## top customers

```python
%sql
SELECT name,SUM(revenue) AS total_revenue FROM fact_sales GROUP BY name ORDER BY total_revenue DESC;
```

## payment method analysis

```python
%sql
SELECT payment_method,COUNT(*) AS total_transactions,SUM(amount) AS total_amount FROM fact_sales GROUP BY payment_method ORDER BY total_amount DESC;
```
